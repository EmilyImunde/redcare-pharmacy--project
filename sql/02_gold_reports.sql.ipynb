```sql
CREATE OR REPLACE TABLE redcare_dev.operations.gold_daily_orders AS
SELECT
   order_day,
   COUNT(*) AS total_orders,
   ROUND(SUM(order_amount), 2) AS total_revenue
FROM redcare_dev.operations.silver_orders
GROUP BY order_day;
```

```sql
CREATE OR REPLACE TABLE redcare_dev.operations.gold_order_status AS
SELECT
    order_day,
    order_status,
    COUNT(*) AS order_count
FROM redcare_dev.operations.silver_orders
GROUP By order_day, order_status
```
